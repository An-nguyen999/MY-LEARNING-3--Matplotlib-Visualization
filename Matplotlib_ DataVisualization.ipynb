# No.1: Pylot API

```python
x = np.linspace(0,10,1000) #chọn một dãy số từ 0 đến 10 có khoảng cách là 1000
plt.plot(x, np.sin(x), color = 'blue', linestyle ='dashed', label = "sin(x)");
plt.plot(x, np.cos(x), color = 'red', label = "cos(x)"); #thêm label phải thêm dòng plt.legend();thì mới hiển thị

plt.title('Sin(x) and Cos(x)');
plt.xlabel("x");
plt.ylabel("F(x)");
#plt.xlim([0,4]); #giới hạn đồ thị tại một khoảng hoặc là
#plt.axis([x min, x max, y min, y max]) -> cú pháp

#plt.axis('tight') #hiển thệ trên đồ thị theo ý mình muốn bung ra hoặc co vào
plt.axis('equal')

plt.legend();
```

# No.2: Object-Oriented (OO) API

```python
#chuẩn bị dữ liệu
x=[1,2,3,4]
y=[2,12,24,35]
```

```python
# gọi hàm
fig, ax = plt.subplots(figsize=(5,5))

#vẽ đồ thị
ax.plot(x,y);

ax.set(title ='A simpe plot', xlabel='x-axis', ylabel = 'y-axis');
```

```python
#c1: pylot API
plt.scatter(x, np.exp(x)); #y= e mũ x
```

```python
#c2: OO API

rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
sizes = 1000*rng.rand(100)
colors = rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s= sizes, c = colors, cmap = 'viridis', alpha = 0.3);

fig.colorbar(img1);
```
