# No.1: Pylot API

```python
x = np.linspace(0,10,1000) #chọn một dãy số từ 0 đến 10 có khoảng cách là 1000
plt.plot(x, np.sin(x), color = 'blue', linestyle ='dashed', label = "sin(x)");
plt.plot(x, np.cos(x), color = 'red', label = "cos(x)"); #thêm label phải thêm dòng plt.legend();thì mới hiển thị

plt.title('Sin(x) and Cos(x)');
plt.xlabel("x");
plt.ylabel("F(x)");
#plt.xlim([0,4]); #giới hạn đồ thị tại một khoảng hoặc là
#plt.axis([x min, x max, y min, y max]) -> cú pháp

#plt.axis('tight') #hiển thệ trên đồ thị theo ý mình muốn bung ra hoặc co vào
plt.axis('equal')

plt.legend();
```

# No2. Scatter

```python
#c1: pylot API
plt.scatter(x, np.exp(x)); #y= e mũ x
```

```python
#c2: OO API

rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
sizes = 1000*rng.rand(100)
colors = rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s= sizes, c = colors, cmap = 'viridis', alpha = 0.3);

fig.colorbar(img1);
```

# No3. Bar
* Vertical
* Horizontal

```python
Loai_nuoc = {'Coke': 10, 'Coca':12, 'Sprite' :30}

fig, ax = plt.subplots()
ax.bar(Loai_nuoc.keys(),Loai_nuoc.values());
ax.set(title ="Bach Hoa Xanh Soft Drink Prices",ylabel = "Prices" );
```

```python
fig, ax = plt.subplots()
ax.barh(list(Loai_nuoc.keys()),list(Loai_nuoc.values())); #hàm barh vẽ ngang này phải list mới ra được đồ thị
ax.set(title ="Bach Hoa Xanh Soft Drink Prices",xlabel = "Prices");
```

# No4. Histogram

```python
np.random.seed(42)
student_height = np.random.normal(170,10,250)
student_height[:10]
```

```python
plt.hist(student_height, bins = 50);
```

# No5. Subplots

```python
#Cách 1:

fig , ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols =2, figsize = (10,5))

#vẽ trên từng subplots khác nhau
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(100), np.random.random(100));
ax3.bar(Loai_nuoc.keys(),Loai_nuoc.values());
ax4.hist(student_height, bins = 50);
```

```python
#Cách 2: thay vì có từng ax1->4 thì thay bằng ax[0,0]...các kiểu

fig , ax = plt.subplots(nrows=2, ncols =2, figsize = (10,5))

#vẽ trên từng subplots khác nhau
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(100), np.random.random(100));
ax[1,0].bar(Loai_nuoc.keys(),Loai_nuoc.values());
ax[1,1].hist(student_height, bins = 50);
```
