```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('C:\\Users\\HP\\Documents\\Python-Jupiter\\Matplotlip - 19.03.2024\\california_cities.csv')
cities.head()
```

```python
cities['area_total_km2'].sort_values(ascending=False) #xem diện tich nó thế nào để phục vụ trang trí biểu đồ
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
lat, lon = cities['latd'],cities['longd'];
pop, area = cities['population_total'],cities['area_total_km2'];
```

```python
plt.figure(figsize =(8,6)); #điều chỉnh size của biểu đồ
plt.style.use('seaborn');
plt.scatter(lon, lat, c = np.log10(pop), cmap = 'viridis', s=area);

plt.axis('equal'); #cho hai trục bằng nhau
plt.xlabel('Longtitude');
plt.ylabel('Latitude');
plt.colorbar(label = 'log$_{10}$(population)');
plt.clim(3,7); #điều chỉnh scale của thanh colorbar

area_range=[50,100,300,500,700]
for area in area_range:
    plt.scatter([],[],s=area, alpha = 0.4,label = str(area)+'km$^2$')
plt.legend(labelspacing = 2, title = "City area");

plt.title('Califonia Cities: Population and Distribution');
```
